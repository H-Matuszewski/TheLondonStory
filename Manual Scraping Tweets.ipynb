```python
import statsmodels.api as sm
X=[Botometer_Scores,Is_repeated_reply,Is_Hindi,Profile_Pic_Has_Indian_Flag,Whataboutism_Country,Whataboutism_Appeal_To_Insignificance,
  Questioning_Legitimacy,Followed_By_Modi_True,Followed_By_AmitShah_True]
def reg_m(y, x):
    ones = np.ones(len(x[0]))
    X = sm.add_constant(np.column_stack((x[0], ones)))
    for ele in x[1:]:
        X = sm.add_constant(np.column_stack((ele, X)))
    results = sm.OLS(y, X,cov_type='HAC').fit()
    return results
print(reg_m(In_Favour, X).summary())
#regression of how other lists affect whether an individual is in favour or not. That's the idea anyway
```

## C - Justifying The Metrics

Measuring bot activity - The tweets were tested to see whether the replies came from actual people or rather were a coordinated effect of spamming replies with bots with the intention to flood out debate or to perhaps over exacerbate a particular point. The scores obtained from Botometer are a decent guess but by no means exact, and the exact determination of whether an account is a bot or not cannot be known. However, the scores are still a sufficient guess for the purpose of this study.<br>
<br>
In Favour - This was measured to get a sense of the general crowd's reaction. These tweets were initially chosen as they were meant to be examples of tweets with highly negative reactions. The mean hostility value seems to show this. Error was made on the side of neutrality (so if a tweet seemed somewhat negative, but was debatable, then it was marked as a 0).<br>
<br>
Is Repeated Reply - The idea here was to see whether there were accounts that were spamming multiple replies, again to perhaps drown out debate. It is possible that a bot account or an account paid to disseminate propaganda could replies multiple times in the hopes that one of the replies gets a lot of likes and is made relevant in the discussion.<br>
<br>
Is Hindi - This was measured mainly to see if the tweets in Hindi had any interesting differences to their English counterparts.<br>
<br>
Profile Pic Is Indian Flag - Having the indian flag as one's profile picture is indicative of some nationalist tendencies, and as such it was interesting to see how this correlated with defending the actions of the indian government, and which particular types of comments came from individuals given that their profile picture was the indian flag. This was counted as either just the flag, or the flag being part of the profile picture alongside other entities (within reason, if it was too small or blurry to be clearly distinguished, then it was marked as 0. Again, erred in favour of neutrality).<br>
<br>
Whataboutism Country - This was measured as appealing to the fact that one is not from the country in question, or doesn't work there (etc etc) could obfuscate from the main point made in the tweet. Mentioning that someone isn't from the country/ runs errands for another is irrelevant, (perhaps just short of a logical fallacy), and as such may be a trait/tactic in coordinated hostile behaviour on twitter.<br>
<br>
Whataboutism Appeal To Insignificance - This was measured as appealing to the fact that the individual isn't all that influential in the grand scheme of the problem being addressed. USually, such tweets tried to imply that the person shouldn't bother talking about it because their voice doesn't matter/won't be heard by the people that matter/ so on. This is again, bordering on a logical fallacy, and may be used to shut certian people down from speaking on a given issue.<br>
<br>
Appeal To Bad Faith - This was measured as tweets which indicated that the tweet was posted for ulterior motives such as financial compensation, internet clout, etc etc. This may be a tactic to stop opposition as this may be used to convince people that the person has no vested interest in the matter, and is thus not contributing to the discussion in any meaningful wasy with a valid opinion. May also be used to argue that the opinion itself is propagated, and perhaps stems from, individuals with ulterior motives.<br>
<br>
Questioning Legitimacy - This is denoted by tweets which seem to attempt to discredit the person being tweeted about (such as Zubair not being a real journalist). This trivialises the person and can make their imprisonment seem more justified.<br>
<br>
Appeal To Ignorance - Denoted by tweets which argue that the original poster isn't sufficiently knowledgeable with the matter at hand to tweet anything about. This also includes tweets which question the intelligence of the original poster. There is room for discussion regarding whether ignorance and stupidity should be bundled together. An appeal to insufficient knowledge/intelligence can be used to devalue someone's point. It generally obfuscates from addressing the point at hand, which could make it a potentially useful coordinated aggressive behaviour.<br>
<br>
Appeal Against EU - Tweets which demonise, or make a stance, against the EU. Usually, the tweets carry a tone of fear/anger/mistrust of foreign entities, such as the EU, interacting with Indian affairs. These are coupled by tweets merely stating the lack of intent to have anything to do with the EU, or that the EU is not needed at all. MEP tweets on Indian affairs can carry a message of desiring coopeartion (such as the one from Alamets√§: "...For the EU to be able to strengthen its cooperation with India, the country must commit..."). Appealing agaisnt the EU usually carries a patriotic undertone, and can be used as a propaganda tool to instoll unity over national pride/national self sufficiency, etc.<br>
<br>
Is Misogyny - A simple metric of whether a tweet is misogynistic or not. This measures explicit misogyny, wherein a tweet will specifically devalue the original post by appealing to the gender of the original poster. This is not to ignore the presence of inplicit misogyny. A more nuanced analysis could compare the replies between male and female replies, and look at either the level of negative posts, the type of tone used in both, whether one contains more appeal to ignornace, and so on.<br>
<br>
Followed by Modi/Shah - With coordinated aggressive behaviour, it has been noted that both of these figures have followed accounts which are specifically tasked with disseminating hate online. Thus a follow from these accounts (while not being definitive proof by any means) could warrant an interesting study as to why the follow arised.<br>
<br>
Whataboutism Someone Else - Denoted by an appeal to someone else being unjustly arrested (unjustly in the eyes of the one which replies). This is meant to discredit the original poster and paint them as partisan.<br>
